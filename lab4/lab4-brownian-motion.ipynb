```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
!ls -l
```

```python
trajectory = pd.read_csv("trajectory-1.csv")
sbn.set()
```

```python
x = trajectory['x']
y = trajectory['y']
```

```python
plt.scatter(x, y, s=1)
```

```python
plt.hist(x, 'auto')
plt.xlabel("horizontal displacement (pixels)")
plt.ylabel("occurences")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
sbn.distplot(x, ax=ax1)
sbn.distplot(y, ax=ax2)
plt.show()
```

```python
sbn.jointplot(x, y, kind='kde')
```
